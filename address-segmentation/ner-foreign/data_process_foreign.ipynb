# Combine and Process foreign addresses
### (Malaysia & Indonesia)

```python
import pandas as pd
```

```python
# combine data from different countries

# read pickle files
df_my = pd.read_pickle("./data/data_collection/my_addresses_w_labels.pkl")
df_id = pd.read_pickle("./data/data_collection/id_addresses_w_labels.pkl")
```

```python
df_combined = pd.concat([df_my, df_id])
df_combined = df_combined.replace(',','', regex=True).dropna(subset=['street']).drop_duplicates(subset=['address']).reset_index(drop=True)
df_combined = df_combined.drop(df_combined[df_combined['street'].str.len() < 15].index).reset_index(drop=True)
df_combined
```

```python
df_combined.to_pickle('./data/data_collection/foreign_combined_w_labels.pkl')
df_combined.to_csv('./data/data_collection/foreign_combined_w_labels.csv', index=False)
```

### Labels for foreign addresses

```python
df_my_collection = pd.read_pickle("./data/data_collection/my_db_collection.pkl")
df_id_collection = pd.read_pickle("./data/data_collection/id_db_collection.pkl")
```

```python
df_collection_combined = pd.concat([df_my_collection, df_id_collection])
df_collection_combined = df_collection_combined.replace(',','', regex=True).dropna(subset=['street']).reset_index(drop=True)
df_collection_combined
```

```python
df_collection_combined.to_pickle('./data/data_collection/foreign_labels.pkl')
df_collection_combined.to_csv('./data/data_collection/foreign_labels.csv', index=False)
```
