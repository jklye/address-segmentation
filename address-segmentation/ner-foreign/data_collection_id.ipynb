### Webscrape address components and create full addresses

```python
def scrape_address(session, url, iteration):
    headers = {
        'User-Agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:52.0) Gecko/20100101 Firefox/52.0'
    }
    try:
        response = session.get(url, headers=headers)
        response.raise_for_status()
        content = response.text

        soup = BeautifulSoup(content, 'html.parser')
        unordered_list = soup.find_all('li', class_='col-sm-6')

        address_batch = []
        street_batch = []
        zipcode_batch = []
        city_batch = []
        province_batch = []
        country_batch = []

        for para in unordered_list:
            lines = para.find_all('span')

            street_line = lines[0].text.upper().replace('STREET:', '').strip()
#             if any(i.isdigit() for i in street_line[:10]):
#                 continue
            city_line = lines[1].text.upper().replace('CITY:', '').strip()
            province_line = lines[2].text.upper().replace('STATE/PROVINCE/AREA:', '').strip()
            zipcode_line = lines[4].text.upper().replace('ZIP CODE', '').strip()
            country_line = lines[6].text.upper().replace('COUNTRY', '').strip()

            street_batch.append(street_line)
            zipcode_batch.append(zipcode_line)
            city_batch.append(city_line)
            province_batch.append(province_line)
            country_batch.append(country_line)

            address = street_line + " " + city_line + " " + province_line + " " + zipcode_line +  " " + country_line
            address_batch.append(address)

        print(f"Iteration: {iteration}")

        return address_batch, street_batch, city_batch, province_batch, zipcode_batch, country_batch

    except requests.exceptions.RequestException as e:
        print(f"Error occurred: {str(e)}")


def scrape_multiple_addresses(url, num_iterations):
    address_list = []
    street_list = []
    zipcode_list = []
    city_list = []
    province_list = []
    country_list = []

    with requests.Session() as session:
        for i in range(num_iterations):
            response = scrape_address(session, f"{url}?quantity=20", i + 1)
            if response:
                address_batch, street_batch, city_batch, province_batch, zipcode_batch, country_batch = response
                address_list.extend(address_batch)
                street_list.extend(street_batch)
                city_list.extend(city_batch)
                province_list.extend(province_batch)
                zipcode_list.extend(zipcode_batch)
                country_list.extend(country_batch)

    return address_list, street_list, city_list, province_list, zipcode_list, country_list
```

```python
url = 'https://www.bestrandoms.com/random-address-in-id'
num_iterations = 20000

address_list, street_list, city_list, province_list, zipcode_list, country_list = scrape_multiple_addresses(
    url, num_iterations
)
```

```python
df = pd.DataFrame(data=list(zip(address_list, street_list, city_list, province_list, zipcode_list, country_list)),
                  columns=['address', 'street', 'city', 'province', 'zipcode', 'country']).dropna(subset=['address']).drop_duplicates(subset=['address'], ignore_index=True).replace(r'^\s*$', np.nan, regex=True)
```

### Preprocessing data from Github

```python
address_list = []

for index, row in df2.iterrows():
    print(f"Processing {index+1} / {len(df2)}")
    
    province = str(row['province']).upper() if pd.notnull(row['province']) and row['province'] != "" else ""
    city = str(row['city_regency']).upper() if pd.notnull(row['city_regency']) and row['city_regency'] != "" else ""
    district = str(row['district']).upper() if pd.notnull(row['district']) and row['district'] != "" else ""
    sub_district = str(row['sub_district']).upper() if pd.notnull(row['sub_district']) and row['sub_district'] != "" else ""
    zipcode = str(row['postal_code']) if pd.notnull(row['postal_code']) and row['postal_code'] != "" else ""
    country = "INDONESIA"
    
    for variation in ["123", "123A", "NO."]:
        if variation == "123":
            house_raw = str(int(np.random.randint(1, 999, size=1)))
            house = house_raw
        elif variation == "123A":
            house_raw = str(int(np.random.randint(1, 999, size=1))) + random.choice(string.ascii_letters).upper()
            house = house_raw
        else:  # "NO."
            house_raw = str(int(np.random.randint(1, 999, size=1)))
            house = "NO. " + house_raw
        
        street = str(district) + " " + str(sub_district) + " " + house
        address = street + " " + city + " " + province + " " + zipcode + " " + country

        address_list.append([address, street, zipcode, city, province, country])
```

### Webscrape from Pos Indonesia

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:52.0) Gecko/20100101 Firefox/52.0'
}

page_range = range(1, 1625)
total_pages = len(page_range)
href_list = []

with requests.Session() as session:
    for page_number in page_range:
        base_url = f'https://idn.youbianku.com'
        page_url = f'{base_url}/location?page={page_number}'

        try:
            response = session.get(page_url, headers=headers)
            response.raise_for_status()

            soup = BeautifulSoup(response.content, 'html.parser')
            contents = soup.find_all('span', class_='field-content')

            for content in contents:
                href = content.find('a')
                if href:
                    href_content = href.get('href')
                    target_url = base_url + href_content
                    href_list.append(target_url)

            print(f"Page: {page_number}/{total_pages}")

        except requests.exceptions.RequestException as e:
            print(f"Error occurred while scraping page {page_number}: {str(e)}")
            break
            
        time.sleep(0.2)
            
print(href_list)
```

```python
address_list = []

total_urls = len(href_list)
for url in tqdm(href_list, desc='Processing URLs', total=total_urls):
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.content, 'html.parser')
    
    rows = soup.find_all('tr')
    for row in rows:
        fields = row.find_all('td')
        if fields:
            street_raw = fields[3].text.strip().upper() + " " + fields[2].text.strip().upper() # location + district
            city = fields[1].text.strip().upper() # regency
            province = fields[0].text.strip().upper() # province
            zipcode = fields[4].text.strip().upper() # postcode
            country = "INDONESIA"
            
            for variation in ["123", "123A", "NO."]:
                if variation == "123":
                    house_raw = str(int(np.random.randint(1, 999, size=1)))
                    house = house_raw
                elif variation == "123A":
                    house_raw = str(int(np.random.randint(1, 999, size=1))) + random.choice(string.ascii_letters).upper()
                    house = house_raw
                else:  # "NO."
                    house_raw = str(int(np.random.randint(1, 999, size=1)))
                    house = "NO. " + house_raw

                street = street_raw + " " + house
                address = street + " " + city + " " + province + " " + zipcode + " " + country

                address_list.append([address, street, city, province, zipcode, country])

df3 = pd.DataFrame(address_list, columns=['address', 'street', 'city', 'province', 'zipcode', 'country'])
df3
```

```python
df3.to_pickle('./data/data_collection/webscraped_id_addresses_2.pkl')
df3.to_csv('./data/data_collection/webscraped_id_addresses_2.csv', index=False)
```
